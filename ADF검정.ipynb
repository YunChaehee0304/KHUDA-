```python
### ADF 단위근 검정

from statsmodels.tsa.stattools import adfuller

# 결과 저장용 리스트
results = []

# DataFrame의 각 열에 대해 반복
for col in dataset.columns:
    ADF_result = adfuller(dataset[col], maxlag=1)
    pd.set_option('display.float_format', '{:.6f}'.format)
    # 결과 저장 (열 이름, 통계량, p값, 유의 수준 임계값들)
    results.append({
        'Column': col,
        'ADF Statistic': ADF_result[0],
        'p-value': ADF_result[1],
        '1% Critical Value': ADF_result[4]['1%'],
        '5% Critical Value': ADF_result[4]['5%'],
        '10% Critical Value': ADF_result[4]['10%'],
        'Is Stationary (p<0.05)': ADF_result[1] < 0.05
    })

# 보기 좋게 DataFrame으로 변환
results = pd.DataFrame(results)
results
```

```python
transform_map = {
    '소비자물가총지수': 'log',
    '매출액': 'log',
    '수출': 'log',
    '연구개발비': 'log',
    '소비자심리지수': 'boxcox',
    '실업률': 'boxcox',
    '가동률지수': 'boxcox',
    '환율': 'log',
    '기준금리': 'sqrt',
    'ROE' : 'log'
}
```

```python
###GPT 변환 코드

import numpy as np
import pandas as pd
from scipy.stats import boxcox

df_adf = dataset.copy()
bc_lambdas = {}

for col, method in transform_map.items():
    s = dataset[col]

    # (1) 결측/0/음수 처리
    if method == 'boxcox':
        # 모두 양수인지 확인
        if (s <= 0).any():
            shift = abs(s.min()) + 1e-6
            s = s + shift
        vals, lam = boxcox(s.dropna())
        bc_lambdas[col] = lam
        # 인덱스 매칭해서 할당
        df_adf[col] = pd.Series(vals, index=s.dropna().index)

    elif method == 'log':
        df_adf[col] = np.log(s.replace(0, np.nan)).fillna(method='bfill')

    elif method == 'sqrt':
        df_adf[col] = np.sqrt(s.clip(lower=0))

    else:  # 원본 유지
        df_adf[col] = s

# 상수열 제거
consts = [c for c in df_adf.columns if df_adf[c].nunique() <= 1]
if consts:
    df_adf.drop(columns=consts, inplace=True)
    print("상수열 제거:", consts)

# 결과 출력
pd.set_option('display.max_columns', None)
print("=== ADF용 변환 데이터 ===")
df_adf
```

```python
### ADF 단위근 검정1

from statsmodels.tsa.stattools import adfuller

# 결과 저장용 리스트
results = []

# DataFrame의 각 열에 대해 반복
for col in df_adf.columns:

    ADF_result = adfuller(df_adf[col], maxlag=1)
    pd.set_option('display.float_format', '{:.6f}'.format)
    # 결과 저장 (열 이름, 통계량, p값, 유의 수준 임계값들)
    results.append({
        'Column': col,
        'ADF Statistic': ADF_result[0],
        'p-value': ADF_result[1],
        '1% Critical Value': ADF_result[4]['1%'],
        '5% Critical Value': ADF_result[4]['5%'],
        '10% Critical Value': ADF_result[4]['10%'],
        'Is Stationary (p<0.05)': ADF_result[1] < 0.05
    })

# 보기 좋게 DataFrame으로 변환
results = pd.DataFrame(results)
results
```

```python
### 1차 차분

# (1) 비정상 변수 리스트
non_stat = results.loc[results['Is Stationary (p<0.05)']==False, 'Column'].tolist()

# (2) df_adf 복사
df_diff1 = df_adf.copy()

# (3) 비정상 변수에만 diff 적용
df_diff1[non_stat] = df_diff1[non_stat].diff()

# (4) 첫 번째 행의 NaN 제거 (선택)
df_diff1 = df_diff1.iloc[1:]

# (5) 결과 확인
print("1차 차분을 적용한 변수:", non_stat)
pd.set_option('display.max_columns', None)
print("=== df_adf 전체 중, 비정상 변수만 1차 차분 적용 결과 ===")
df_diff1
```

```python
df_diff1 = df_diff1.reset_index()          # '연도'라는 새 칼럼이 생깁니다😊
cols_to_select = [
     '분기','경제성장률(전기대비)', '연구개발비','ROE','매출액','가동률지수'
]
df_diff1 = df_diff1[cols_to_select].copy()
```

```python
df_diff1 = df_diff1.set_index('분기')                 # 연도를 인덱스로
df_diff1.index = pd.PeriodIndex(
    df_diff1.index.astype(str), freq='Q'            # 연 단위 주기 설정😊
)
```

```python
from statsmodels.tsa.api import VAR

# df: (T × k) 형태의 시계열 DataFrame
model = VAR(df_diff1)  
# 최대 10차수까지 정보기준 계산
lag_order_results = model.select_order(maxlags=7)  

print(lag_order_results.summary())
```

```python
# (예시) df: (T × k) 형태의 시계열 DataFrame
# df = pd.read_csv('your_data.csv', index_col=0)

# 1) VAR 모델 객체 생성
var_mod = VAR(df_diff1)

# 2) lag 0으로 적합
#    trend='c' → 상수항 포함, 필요 없으면 'n'
var_res = var_mod.fit(
    maxlags=3,   # 또는 fit(0) 도 가능
    trend='c'    # 'c': constant, 'nc': no constant
)

print(var_res.summary())   # 계수 · R² 등 확인
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant
dataset
```

```python

def calculate_vif(X):
    X = add_constant(X)  # 상수항 추가
    vif_data = pd.DataFrame()
    vif_data["변수명"] = X.columns
    vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    return vif_data

# 사용 예시
# df = pd.read_csv("your_data.csv")
# vif_result = calculate_vif(df[['x1', 'x2', 'x3']])
# print(vif_result)
vif_result = calculate_vif(dataset)
print(vif_result)

corr = dataset.corr()


threshold = 0.8

def highlight_high(val):
    """
    절댓값이 threshold 이상이면 배경색, 아니면 공백 반환
    """
    return 'background-color: yellow' if abs(val) >= threshold else ''

# Styler 객체에 applymap으로 함수 적용
styled_corr = corr.style.map(highlight_high)

# Jupyter 환경이라면 이 한 줄만으로 결과가 표시됩니다.
styled_corr
```
